```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
netflix_titles = "Resources/netflix_titles.csv"
```

```python
netflix_df = pd.read_csv(netflix_titles)
netflix_df.head()
```

```python
netflix_tv = netflix_df.loc[netflix_df["type"] == "TV Show",:]
```

```python
netflix_tv.head()
```

```python
# netflix_new = netflix_tv.drop(netflix_tv.columns[["show_id","director"]], axis=1)
netflix_new = netflix_tv[["type","title","country","release_year","listed_in"]]
```

```python
netflix_new.head()
```

```python
netflix_new.nunique()["title"]
```

```python
netflix_shows = netflix_new.drop_duplicates("title")
print(netflix_shows.count())
netflix_shows.head()
```

```python
# netflix_shows.to_csv('netflix_shows',index=False)
netflix_shows.rename(columns={"listed_in":"category"})
```

```python
netflix_shows.to_csv('Resources/netflix_shows',index=False)
```

```python
connection_string = "postgres:60221023@localhost:5432/netflix_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
engine.table_names()
```

```python
netflix_shows.to_sql(name='netflixshows', con=engine, if_exists='append', index=True)
```
